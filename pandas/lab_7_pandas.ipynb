**Прочитать файл CSV и перевести каждую 100-ую строку в DataFrame**

указать chunksize, возвращаемое значение будет итеративным объектом типа TextFileReader:  
iloc - Индексирование исключительно на основе целочисленных позиций для выбора по позиции.

```python
a = pd.read_csv('https://raw.githubusercontent.com/Grossmend/CSV/master/titanic/data.csv', chunksize = 100)
b = pd.DataFrame()
for elem in a:
    b = b.append(elem.iloc[0,:])
b
```
